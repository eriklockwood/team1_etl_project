```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
from secret import username, password
```

```python
# Import IMDB Movies data
imdb_file = "../etl_project_data/IMDb movies.csv"
imdb_movies_df = pd.read_csv(imdb_file)
imdb_movies_df.head()
```

```python
imdb_movies_df.sort_values(by=['year'], ascending=True, inplace=True)
imdb_movies_df = imdb_movies_df[~(imdb_movies_df['year'] < 2000)]
imdb_movies_df.info()
```

```python
select_imdb_movies_df = imdb_movies_df[['title', 'year', 'duration', 'budget', 'worlwide_gross_income']]
select_imdb_movies_df.head(50)
```

```python
select_imdb_movies_df['movie_id'] = range(1, len(select_imdb_movies_df)+1)
select_imdb_movies_df.set_index('movie_id', inplace=True)
select_imdb_movies_df.tail()
```

```python
# Import Streaming availability data
stream_file = "../etl_project_data/MoviesOnStreamingPlatforms_updated.csv"
stream_df = pd.read_csv(stream_file)
stream_df.head()
```

```python
stream_df.sort_values(by=['Year'], ascending=True, inplace=True)
stream_df = stream_df[~(stream_df['Year'] < 2000)]
stream_df.head()
```

```python
stream_select_df = stream_df[['Title', 'Year', 'Netflix', 'Hulu', 'Prime Video', 'Disney+']]
```

```python
stream_select_df.rename(columns={'Title':'title', 'Year':'year', 'Netflix':'netflix', 'Hulu':'hulu', 'Prime Video':'prime', 'Disney+':'disney'}, inplace=True)
stream_select_df.keys()
```

```python
stream_select_df['joinstring'] = stream_select_df['title'].str.lower() + stream_select_df['year'].apply(str)
stream_select_df['joinstring'] = stream_select_df['joinstring'].str.replace(" ","",)
stream_select_df.head()
```

```python
stream_select_df = stream_select_df[['netflix', 'hulu', 'prime', 'disney', 'joinstring']]
```

```python
select_imdb_movies_df['joinstring'] = select_imdb_movies_df['title'].str.lower() + select_imdb_movies_df['year'].apply(str)
select_imdb_movies_df['joinstring'] = select_imdb_movies_df['joinstring'].str.replace(" ","",)
select_imdb_movies_df.head()
```

```python
merge_df = pd.merge(left=select_imdb_movies_df, right=stream_select_df, how='left', left_on='joinstring', right_on='joinstring')
merge_df.head()
```

```python
merge_df = merge_df[['title', 'year', 'duration', 'budget', 'worlwide_gross_income', 'netflix', 'hulu', 'prime', 'disney']]
merge_df['netflix'] = merge_df['netflix'].fillna(0)
merge_df['hulu'] = merge_df['hulu'].fillna(0)
merge_df['prime'] = merge_df['prime'].fillna(0)
merge_df['disney'] = merge_df['disney'].fillna(0)
merge_df.head()
```

```python
merge_df['netflix'].replace(to_replace=0, value=False, inplace=True)
merge_df['netflix'].replace(to_replace=1, value=True, inplace=True)
merge_df['hulu'].replace(to_replace=0, value=False, inplace=True)
merge_df['hulu'].replace(to_replace=1, value=True, inplace=True)
merge_df['prime'].replace(to_replace=0, value=False, inplace=True)
merge_df['prime'].replace(to_replace=1, value=True, inplace=True)
merge_df['disney'].replace(to_replace=0, value=False, inplace=True)
merge_df['disney'].replace(to_replace=1, value=True, inplace=True)

merge_df.head()
```

```python
merge_df['budget'] = merge_df['budget'].str.replace('[^\d\.]', '').astype(float)


merge_df.head()
```

```python
merge_df['worlwide_gross_income'] = merge_df['worlwide_gross_income'].str.replace('[^\d\.]', '').astype(float)
merge_df.head()
```

```python
merge_df['movie_id'] = range(1, len(merge_df)+1)
merge_df.head()
```

```python
rds_connection_string = f'{username}:{password}@localhost:5432/etl_project'
engine = create_engine(f'postgresql+psycopg2://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
merge_df.to_sql(name='movie', con=engine, if_exists='append', index=False)
```
