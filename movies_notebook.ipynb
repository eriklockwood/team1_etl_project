```python
import pandas as pd
from sqlalchemy import create_engine
from secret import username, password
```

```python
# Import IMDB Movies data
xlsx_file = "../etl_project_data/IMBD Movies.xlsx"
imdb_movies_df = pd.read_excel(xlsx_file)
imdb_movies_df.head()
```

```python
imdb_movies_df.keys()
```

```python
imdb_movies_df.info()
```

```python
imdb_movies_df.sort_values(by=['year'], ascending=True, inplace=True)
imdb_movies_df = imdb_movies_df[~(imdb_movies_df['year'] < 2000)]
imdb_movies_df.info()
```

```python
select_imdb_movies_df = imdb_movies_df[['title', 'year', 'duration', 'budget', 'worlwide_gross_income']]
select_imdb_movies_df.head(50)
```

```python
select_imdb_movies_df['movie_id'] = range(1, len(select_imdb_movies_df)+1)
select_imdb_movies_df.set_index('movie_id', inplace=True)
select_imdb_movies_df.tail()
```
